1  Explain the difference between Relative, Absolute and Mixed Cell Referencing.

  - Relative cell references change automatically when a formula is copied to another cell, maintaining the relative position between cells.
    
    Example: if you copy the formula =B2/A2 from the cell "C2" to "C3",the formula in "C3" will automatically become =B3/A3
  - Absolute cell references remain fixed no matter where the formula is copied.To get an absolute reference from a relative,we can use the dollar sign($) character

    Example: if G2 = C2/$A$2,when copied to G3, G3 becomes = C3/$A$2
    
  - Mixed cell reference is a combination of relative and absolute cell reference.The dollar($) before the row number fixes the row and before the column name fixes the column.

    Example:$A1 -fixed A Column, Row changes. A$1 Row 1 fixed, Column changes.
